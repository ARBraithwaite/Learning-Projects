# Feature Engineering

```python
# Donations per month

merged['Donations per Month'] = round(merged['Number of Donations'] / merged['Months since First Donation'], 2)
```

```python
# Volume of blood donated per visit
merged['Average Volume Donated per Visit'] = round(merged['Total Volume Donated (c.c.)'] / merged['Number of Donations'], 2)
```

```python
# VIF graphs to identify multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

iv_feats = merged.drop(['Made Donation in March 2007','Average Volume Donated per Visit'], axis = 1)

vif = []
for i in range(1,len(add_constant(iv_feats).columns)):
    vif.append(variance_inflation_factor(add_constant(iv_feats).values, i))
    
vif = np.array(vif)
```

```python
sns.set(font_scale = 1)
f, ax = plt.subplots(figsize=(15,10))
sns.barplot(iv_feats.columns.values, vif)
ax.set_ylabel('VIF')
plt.xticks(rotation=90)
plt.show()
```

# Model 

```python
# Shuffle the data
train_df = merged[:trainlen]

train_df = train_df.sample(frac=1, random_state=RANDOM_STATE)
train_df = train_df.sample(frac=1, random_state=RANDOM_STATE)
train_df = train_df.sample(frac=1, random_state=RANDOM_STATE)
```

```python
test_df = merged[trainlen:]
```

```python
from sklearn.model_selection import train_test_split

X = train_df.drop(['Made Donation in March 2007','Number of Donations','Average Volume Donated per Visit'], axis = 1)
y = train_df['Made Donation in March 2007']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, stratify=y, random_state=RANDOM_STATE)
```

```python
test_df.drop(['Made Donation in March 2007','Number of Donations','Average Volume Donated per Visit'], axis = 1, inplace = True)
```

```python
y = y.astype(np.int64)
np.bincount(y)*100/len(y)
```

```python
import statsmodels.api as sm
logit_model=sm.Logit(y,X)
result=logit_model.fit()
print(result.summary2())
```

```python
predicted = result.predict(X_test)
predicted_choice = (predicted > 0.4).astype(int)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

model = Pipeline([('scaler', StandardScaler()),
                  ('polynomial_features', PolynomialFeatures(degree=3)),
                  ('logit' , LogisticRegression(C=1e9))])
```

```python
model.fit(X_train, y_train)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(model.score(X_test, y_test)))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X_train, y_train, cv=6)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

# Model Evaluation

```python
from sklearn.metrics import confusion_matrix

y_pred = model.predict(X_test)
confusion = confusion_matrix(y_test, y_pred)

fig, ax = plt.subplots(figsize=(10,6))
ax = sns.heatmap(confusion, annot=True, fmt="d", linewidths=0.5)
ax.set(xlabel='Predicted', ylabel='Actual')
ax.set_title('Logit Classifier')
```

```python
from sklearn.metrics import confusion_matrix, auc, precision_recall_curve, roc_curve, log_loss, classification_report

print(classification_report(y_test, y_pred))
```

```python
log_loss(y_test, y_pred)
```

```python
# Compute fpr, tpr, thresholds and roc auc
probs = model.predict_proba(X_test)
y_score = probs[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_score)
roc_auc = auc(fpr, tpr)
```

```python
plt.figure(figsize=(8,8))
plt.plot(fpr, tpr, label='Logistic Regression ROC curve (area = %0.3f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')  # random predictions curve
plt.xlim([-0.01, 1.0])
plt.ylim([0.0, 1.01])
plt.xlabel('False Positive Rate or (1 - Specifity)')
plt.ylabel('True Positive Rate or (Sensitivity)')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
```

```python
precision, recall, _ = precision_recall_curve(y_test, y_score)
```

```python
plt.figure(figsize=(8,8))
plt.plot(recall, precision, label='Logistic Regression')
plt.xlim([-0.01, 1.01])
plt.ylim([0.0, 1.01])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-recall curve, Logit model')
plt.legend(loc="lower right")
```
