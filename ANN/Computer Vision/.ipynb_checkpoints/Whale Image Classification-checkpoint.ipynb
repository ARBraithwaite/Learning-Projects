```python
def load_images(data: list, classes: list):
    """
    """
    for i, image_gen in enumerate(data):
        image_list = [image for image in image_gen]
        if i == 0:
            frame = pd.DataFrame(data={'image features': image_list, 'class': [classes[i]]*len(image_list)})
        else:
            frame = frame.append(pd.DataFrame(
                          data={'image features': image_list,
                                'class': [classes[i]]*len(image_list)
                               }
                         ),
             ignore_index=True)
            
    return frame
```

## Colour Space

For X-ray data gray scale is more suited

It's easier to represent colour in the hsv space compared to rgb.

```python
hsv = cv.cvtColor(train.iloc[0,0], cv.COLOR_BGR2HSV)
gray = cv.cvtColor(train.iloc[0,0], cv.COLOR_BGR2GRAY)
```

```python
fig, axis = plt.subplots(1, 2, figsize=(10,5))

images = [hsv, gray]
titles = ['hsv', 'Gray']

for i, ax in enumerate(axis.ravel()):
    ax.imshow(images[i])
    ax.axis('off')
    ax.set_title(titles[i])
```

```python
def to_grayscale(bgr_space):
    return cv.cvtColor(bgr_space, cv.COLOR_BGR2GRAY)
```
