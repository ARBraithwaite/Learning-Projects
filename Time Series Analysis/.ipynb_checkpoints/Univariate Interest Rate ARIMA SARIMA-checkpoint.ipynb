## Linear/Polynomial Regression

```python
X = [i for i in range(0, len(data))]
X = np.reshape(X, (len(X), 1))
y = data.values

lr_trend, ply_trend = reg_trend(X, y, 5)
trends = [lr_trend, ply_trend]
labels = ['Linear_Trend', 'Polynomial_Trend']

fig, axis = plt.subplots(2, 1, figsize = (15,10))
fig.suptitle('Regression Trends', fontsize=16)

for i in range(len(axis)):
    axis[i].plot(y)
    axis[i].plot(trends[i], label = labels[i])
    axis[i].legend()
```

## Local Linear/Polynomial
For degree=0 it reduces to a weighted moving average
</br>

Having a wider kernel and including more datapoints lowers the noise (variance) but increases the bias as the regression will not be able to capture variations on a scale much narrower than the kernel window. (width)

```python
from localreg import *

X = [i for i in range(0, len(y))]
X = np.array(X)
y = data.values.flatten()

local_reg = localreg(X, y, degree=3, kernel=tricube, frac=0.2)

fig, ax = plt.subplots(1, 1, figsize=(20,10))
ax.plot(y)
ax.plot(local_reg)
ax.set_title('Local Polynomial Regression')
labels=['Ground Truth','Local Polynomial']
ax.legend(labels=labels)
```

## Locally Weighted Scatterplot Smoothing

```python
import statsmodels.api as sm

X = [i for i in range(0, len(y))]
X = np.array(X)
y = data.values.flatten()

lowess_sm = sm.nonparametric.lowess
yest_sm = lowess_sm(y,X,frac=0.1, return_sorted = False)

fig, ax = plt.subplots(1, 1, figsize=(20,10))
ax.plot(y)
ax.plot(yest_sm)
ax.set_title('Trend Smoothing')
labels=['Ground Truth','Lowess']
ax.legend(labels=labels)
```

# Time Series Decomposition
TS Components:
</br>

Level: The average value in the series.
</br>

Trend: The increasing or decreasing value in the series.
</br>

Seasonality: The repeating short-term cycle in the series.
</br>

Noise: The random variation in the series.

#### Additive Model
y(t) = Level + Trend + Seasonality + Noise
#### Multiplicative Model

y(t) = Level x Trend x Seasonality x Noise

## Classical Decomposition

```python
decompose(data['INT'], 'additive')
```

## STL Decomposition
With daily observations and large annual cycles, period=365. For hourly observations with large daily cycles, period=24. Some inspection, and trial and error may be helpful.

```python
stl_decompose(data['INT'], period=15)
```

# Stationarity
If a trend and/or seasonality component is found remove the corresponding estimations
</br>

Use the residual process taken from the time series decomposition
</br>

Differencing the Series until stationary, Take the log of the series and/or Take the nth root of the series
</br>

If count data take sqrt(count+3/4)

```python
data['difference'] = data['INT'] - data['INT'].shift(1) # 1st order difference
data['detrend_smooth'] = data['INT'] - yest_sm
data['detrend_local_reg'] = data['INT'] - local_reg
data['deseason'] = data['INT'] - data['INT'].shift(15)
```

# Test For Stationarity
Mean and variance do change over time
</br>

Doesn't have any trend or seasonal effects
</br>

Plot Distribution, Compare summary statistics, ADF test, plot ACF/PACF

# ACF/PACF Plot
Autocorrelation function (ACF) at lag k, this is the correlation between series values that are k intervals apart.
</br>

A partial autocorrelation (PACF) is the amount of correlation between a variable and a lag of itself that is not explained by correlations at all lower-order-lags.

```python
# Distribution should be normal
stationary_ts = ['difference','detrend_smooth','detrend_local_reg', 'deseason']
for ts in stationary_ts:
    plot_ts(data[ts].dropna(), lags=50)
```

```python
# First order difference the deseasoned data
data['stationary'] = data['deseason'] - data['deseason'].shift(1)
plot_ts(data['stationary'].dropna(), lags = 50)
```

### Split and Compare Summary Statistics

```python
stationary_ts.append('stationary')
for ts in stationary_ts:
    X = data[ts].dropna()
    split = round(len(X) * 0.5)
    X1, X2 = X[0:split], X[split:]
    mean = []
    var = []
    mean.append('{:.5f}'.format(X1.mean()))
    mean.append('{:.5f}'.format(X2.mean()))
    var.append('{:.5f}'.format(X1.var()))
    var.append('{:.5f}'.format(X2.var()))
    print('\n' + ts)
    print(pd.DataFrame([mean,var], index=['mean','variance'], columns=['Part 1','Part 2']))
```

# ADF Test
Augmented Dickey-Fuller test
The null hypothesis of the test is that the time series can be represented by a unit root, that it is not stationary
The alternate hypothesis (rejecting the null hypothesis) is that the time series is stationary.
</br>

A p-value below a threshold (such as 5% or 1%) suggests we reject the null hypothesis (stationary), otherwise a p-value above the threshold suggests we fail to reject the null hypothesis (non-stationary).

```python
for ts in stationary_ts:
    test_stationarity(data[ts].dropna())
```

# Time Series Model
#### Autoregressive Integrated Moving Average (ARIMA)
Made up of - Autoregression (AR) - Model parameter p - suitable for univariate ts without trend and seasonal components.
</br>

Moving Average (MA) - Model parameter q - suitable for univariate time series without trend and seasonal components.
</br>

ARIMA combines both with the addition of parameter d -differencing pre-processing step - suitable for univariate time series with trend and without seasonal components.
</br>

#### Seasonal Autoregressive Integrated Moving-Average (SARIMA)
SARIMA(p, d, q)(P, D, Q)m where m is the number of time steps in each season (the seasonal period).
</br>
#### ARIMA/SARIMA with Exogenous Regressors
Extension - includes modeling of exogenous variables, which are parallel input sequences that have observations at the same time steps as the original series.
#### Vector Autoregression (VAR)
#### Generalized Autoregressive Conditionally Heteroskedastic (GARCH)
Modelling conditional Variance - heteroskedasticity
#### Facebook Prophet
#### LSTM/CNN
#### Simple Exponential Smoothing (SES) 
models the next time step as an exponentially weighted linear function of observations at prior time steps.
</br>

#### Holt Winters Exponential Smoothing (HWES)
Triple Exponential Smoothing method models the next time step as an exponentially weighted linear function of observations at prior time steps, taking trends and seasonality into account.

## ARIMA - SARIMA 
Compare AIC values - lowest value prefered (ignore absolute values)
</br>

P-Value (p>|z|) should be at least less than 0.05 for the respective X to be significant
</br>

Seasonal parameters will be seen in the seasonal lags of the PACF and ACF
</br>

Q look for spike a seasonal lags in the ACF
</br>

P look for spike a seasonal lags in the PACF

## Parameters
<B>p</B> 2, since its the last significant lag on PACF after which most others are becoming not significant
</br>

<B>d</B> 1, first order difference
</br>

<B>q</B> 1  from ACF
</br>

<B>P</B> May be 1, significant spikes at 10th lag on PACF
</br>

<B>D</B> 1  seasonal difference performed
</br>

<B>Q</B> 1, 5th lag is significant

```python
from itertools import product 
# setting initial values and some bounds for them
p = range(0, 5)
d = 1 
q = range(0, 5)
P = range(1, 2)
D = 1 
Q = range(1, 2)
s = 3 # season length

# creating list with all the possible combinations of parameters
arima_parameters = list(product(p, q))
sarima_parameters = list(product(p, q, P, Q))
print(F"Parameter Combinations - ARIMA: {len(arima_parameters)} SARIMA: {len(sarima_parameters)}")
```

```python
result_table_sarima = optimize_SARIMA_ARIMA(data['INT'], sarima_parameters, d, D, s, 'SARIMA')
result_table_sarima[:5]
```

```python
result_table_arima = optimize_SARIMA_ARIMA(data['INT'], arima_parameters, d, D, s, 'ARIMA')
result_table_arima[:5]
```

```python
#ARIMA
p, q = result_table_arima.parameters[0]
arima_model = SARIMAX(data['INT'], order=(p, d, q)).fit(disp=-1)

#SARIMA
p, q, P, Q = result_table_sarima.parameters[0]
sarima_model = SARIMAX(data['INT'], order=(p, d, q), seasonal_order=(P, D, Q, s)).fit(disp=-1)
```

## Ljung Box Test

Test for the absence of serial autocorrelation, up to a specified lag k.
If p-value < 0.05: You can reject the null hypothesis. So you can assume that your values are showing dependence on each other i.e. there is evidence of auto correlation.

```python
from statsmodels.stats.diagnostic import acorr_ljungbox
_, pvalue = acorr_ljungbox(sarima_model.resid, lags=6, boxpierce=False)
list(pvalue)
```

## Error Measures

Mean Absolute Percentage Error (MAPE) - Percentage error
MAPE also cannot be used when the time series can take zero values.
</br>

MAE - The mean absolute error, or MAE, is calculated as the average of the forecast error values.
</br>

MSE/RMSE, is calculated as the average of the squared forecast error values.
MSE/RMSE is scale-dependent. Can't be used for comparing accuracy across time series with different scales.

PM (Performance Measure) - Measure the proportion of the variability in the prediction and the observed data - closer to 0 the better

#### X% MAPE implies the model is about 100-X% accurate in predicting the next N observations.
