# Facebook Prophet

Time series model with three main model components: trend, seasonality, and holidays.
</br>

y(t) = g(t) + s(t) + h(t) + εt

<ul>
  <li>g(t): piecewise linear or logistic growth curve for modelling non-periodic changes in time series</li>
  <li>s(t): periodic changes (e.g. weekly/yearly seasonality)</li>
  <li>h(t): effects of holidays (user provided) with irregular schedules</li>
  <li>εt: error term accounts for any unusual changes not accommodated by the model</li>
</ul>

```python
# Rename columns to ds, y for prophet
data = data.reset_index().rename(columns={'DATE':'ds','INT':'y'})
```

### Modeling Holidays and Special Events
Create a DataFrame with columns holiday and ds - optional columns lower_window/upper_window which can be used to extend holidays n number of days in front and/or prior.
<code>lower_window = n, upper_window = n </code>

```python
# Modelling recession July 1981 and ended in November 1982
recession = pd.DataFrame({
  'holiday': 'recession',
  'ds': pd.to_datetime(['1981-07-01', '1982-11-01']),
})

holidays = recession
```

```python
from fbprophet import Prophet

# Model fit
model = Prophet(growth='linear', weekly_seasonality=False, daily_seasonality=False, yearly_seasonality=False, \
                holidays=holidays, seasonality_mode='multiplicative', n_changepoints=4)

model.add_country_holidays(country_name='UK')
model.add_seasonality(name='yearly', period=5, fourier_order=25)
model.fit(data)

#Forecast
future = model.make_future_dataframe(periods=5, freq='Y')
forecast = model.predict(future)

# PLots
model.plot(forecast)
plt.show()
```

### Changepoint Plot

```python
from fbprophet.plot import add_changepoints_to_plot

fig = model.plot(forecast)
a = add_changepoints_to_plot(fig.gca(), model, forecast)
```

### Cross Validation
Here we do cross-validation to assess prediction performance on a horizon of 1080 days, starting with 18000 days of training data in the first cutoff and then making predictions every 180 days.

```python
from fbprophet.diagnostics import cross_validation
cv = cross_validation(model, initial='18250 days', period='1095 days', horizon = '3650 days')
```

```python
cv.head()
```

rolling_window, specifies the proportion of forecasts to use in each rolling window. The default is 0.1, corresponding to 10% of rows included in each window 

```python
# Performance metrics
from fbprophet.diagnostics import performance_metrics
pm = performance_metrics(cv, rolling_window=0.5)
pm.head()
```

```python
# Plot cv metrics
from fbprophet.plot import plot_cross_validation_metric
fig = plot_cross_validation_metric(cv, metric='mape', rolling_window=0.5)
```
